# Pandas Series

```python
import pandas as pd
import numpy as np
```

```python
# A series is a 1D array which store data values along with idx or labels

# basic examplpe
arr = np.array([10,20,30,40,50])
df = pd.Series(arr)
print(df)
```

```python
# series using labels
arr = np.array([10,20,30,40,50])
labels = ['a','b','c','d','e']
df = pd.Series(arr, index=labels)
print(df)
```
