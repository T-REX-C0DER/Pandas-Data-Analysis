```python
import numpy as np
import pandas as pd
```

# Merging Data

```python
# Data set for merging
student_data = {
    "ID": [101, 102, 103, 104, 105, 106, 107, 108],
    "Name": ["Amit", "Neha", "Ravi", "Pooja", "Kiran", "Sneha", "Arjun", "Meena"],
    "Branch": ["AI", "DS", "AI", "CS", "DS", "CS", "AI", "DS"],
    "Semester": [1, 1, 2, 2, 1, 2, 1, 2],
    "Marks": [78, 85, np.nan, 92, 66, 88, 74, np.nan],
    "Attendance (%)": [90, 95, 88, 98, 75, 92, np.nan, 85]
}
df1 = pd.DataFrame(student_data)

sales_data = {
    "ID": [201, 202, 203, 204, 205, 206, 207, 208],
    "Product": ["Laptop", "Mouse", "Keyboard", "Laptop", "Monitor", "Mouse", "Laptop", "Keyboard"],
    "Category": ["Electronics", "Accessories", "Accessories", "Electronics", "Electronics", "Accessories", "Electronics", "Accessories"],
    "Quantity": [1, 2, 1, 1, 2, 3, 1, 2],
    "Price": [65000, 800, 1500, 70000, 12000, 850, 68000, 1600],
    "Discount (%)": [10, 5, 0, 15, np.nan, 5, 10, 0]
}
df2 = pd.DataFrame(sales_data)
```

```python
pd.merge(df1,df2,on='ID',how='inner')
```

```python
pd.merge(df1,df2,on='ID',how='outer')
```

```python
pd.merge(df1,df2,on='ID',how='right')
```

```python
pd.merge(df1,df2,on='ID',how='left')
```

# Joining Data

```python
# Data set for the joining 
employees = pd.DataFrame({
    'emp_id': [1, 2, 3],
    'name': ['A', 'B', 'C']
})

salaries = pd.DataFrame({
    'emp_id': [1, 2, 4],
    'salary': [50000, 60000, 70000]
})
```

```python
# Types of joining
pd.merge(employees, salaries, on='emp_id', how='inner')
```

```python
pd.merge(employees, salaries, on='emp_id', how='outer')
```

```python
pd.merge(employees, salaries, on='emp_id', how='left')
```

```python
pd.merge(employees, salaries, on='emp_id', how='right')
```
