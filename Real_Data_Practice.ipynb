```python
# Creating new column

def extract_episodes(txt):
    check = False 
    data = ""

    for i in txt :
        if i == ')' :
            check = False
            return data 

        if check == True :
            data += i
            
        if i == '(':
            check = True

    return data 

df["Episodes"] = df["Title"].apply(extract_episodes)
print(df)
```

```python
# Replacing colm data
df["Episodes"] = df["Episodes"].str.replace(" eps" , "")
print(df)
```

```python
# Converting into int

df["Episodes"] = df["Episodes"].astype(int)
print(df)
```

```python
# Creating a column for timestamp

def extract_time(txt) :
    check = False
    data = ""
    for i in range(len(txt)) :
        if txt[i] ==')':
            for j in range(i+1 , i+20) :
                data += txt[j]
            return data

df["Time"] = df["Title"].apply(extract_time)

df.head()
```
